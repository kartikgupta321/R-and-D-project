```python
# import pandas with shortcut 'pd'
import pandas as pd  
from matplotlib import pyplot as plt

# Set the figure size
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

# Make a list of columns
columns = ['i']

# read_csv function which is used to read the required CSV file
data = pd.read_csv('samples.csv')
#   iii,avr,avl,avf,v1,v2,v3,v4,v5,v6,vx,,vz
# display 
print("Original 'input.csv' CSV Data: \n")
print(data)
  
# drop function which is used in removing or deleting rows or columns from the CSV files
data.drop('ii', inplace=True, axis=1)
data.drop('iii', inplace=True, axis=1)
data.drop('avr', inplace=True, axis=1)
data.drop('avl', inplace=True, axis=1)
data.drop('avf', inplace=True, axis=1)
data.drop('v1', inplace=True, axis=1)
data.drop('v2', inplace=True, axis=1)
data.drop('v3', inplace=True, axis=1)
data.drop('v4', inplace=True, axis=1)
data.drop('v5', inplace=True, axis=1)
data.drop('v6', inplace=True, axis=1)
data.drop('vx', inplace=True, axis=1)
data.drop('vy', inplace=True, axis=1)
data.drop('vz', inplace=True, axis=1)
  
# display 
print("\nCSV Data after deleting the column 'year':\n")
print(data)

# Plot the lines
data.plot()
plt.show()
```
