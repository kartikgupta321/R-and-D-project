```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('samples.csv')
data.drop('Elapsed time', inplace=True, axis=1)
data.drop('vx', inplace=True, axis=1)
data.drop('vy', inplace=True, axis=1)
data.drop('vz', inplace=True, axis=1)
data = data.iloc[1:]
data.head()
```

```python
fig,axes = plt.subplots(16,1,figsize =(12,20))
for i, ax in enumerate(axes.flatten()):
    y = data.iloc[:,1].values
    print(y)
    ax.plot(y,color = plt.cm.Paired(i/10.))
plt.show()
```
